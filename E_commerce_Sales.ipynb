```python
import pandas as pd
import numpy as np
from google.colab import files
import seaborn as sns
import matplotlib.pyplot as plt

# Upload the file to the current session
uploaded = files.upload()

# Get the filename from the uploaded dictionary
filename = list(uploaded.keys())[0]  # Assuming only one file is uploaded

# Read the CSV file using the uploaded filename
dataset = pd.read_csv(filename)
```

```python
dataset.isnull().sum().sum()
```

```python
missing_rows = dataset[dataset.isnull().any(axis=1)]
print(missing_rows)
```

convert order_date to datetiem format

```python
dataset['order_date'] = pd.to_datetime(dataset['order_date'])
```

check basic info about the dataset

```python
dataset.info()
```

```python
dataset.head(5)
```

Total revenue in teh dataset

```python
Total_revenue = dataset['revenue'].sum()
print(Total_revenue)
```

Revenue by Month

```python
monthly_revenue = dataset.resample('M', on='order_date')['revenue'].sum()
print(monthly_revenue)
```

```python
monthly_revenue.plot(kind='bar', figsize=(6, 4))
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Monthly Revenue')
plt.show()
```

Top 10 best- selling products by quantity

```python
top_products = dataset.groupby('sku')['quantity'].sum().nlargest(10)
print(top_products)
```

Revenue by product size

```python
size_product = dataset.groupby('size')['revenue'].sum()
print(size_product)
```

Analyze Revenue by product color

```python
color_revernue= dataset.groupby('color')['revenue'].sum()
print(color_revernue)
```

visulize it using bar chart

```python
color_revernue.plot(kind='bar', figsize=(6, 4))
plt.xlabel('Color')
plt.ylabel('Revenue')
plt.title('Revenue by Product Color')
plt.show()
```

Find the most common sizes sold

```python
size_quantity = dataset['size'].value_counts()
print(size_quantity)
```

```python
size_quantity.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6), title="Size Distribution")
plt.ylabel('')  # Removes the y-axis label
plt.show()
```

visulize with a pie chart

```python
size_quantity.plot(kind='barh', figsize=(8, 5), color='skyblue', title="Most Sold Sizes")
plt.xlabel('Quantity Sold')
plt.ylabel('Size')
plt.grid(axis='x')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.barplot(x=size_quantity.values, y=size_quantity.index, palette="viridis")
plt.title("Most Sold Sizes")
plt.xlabel("Quantity Sold")
plt.ylabel("Size")
plt.show()
```

Analyze daily order counts

```python
daily_orders = dataset.resample('D', on='order_date')['order_id'].count()
print(daily_orders)
```

```python
daily_orders.plot(kind='line', figsize=(10, 6))
plt.xlabel('Date')
plt.ylabel('Order Count')
plt.title('Daily Order Counts')
plt.show()
```

find duplicate rows

```python
duplicates = dataset [dataset.duplicated()]
print(f"Number of duplicate rows: {len(duplicates)}")
duplicates.head()
```

Detect high-revenue oultliers

```python
Q1 = dataset['revenue'].quantile(0.25)
Q3 = dataset['revenue'].quantile(0.75)
IQR = Q3 - Q1

# Define upper bound for outliers
upper_bound = Q3 + 1.5 * IQR

# Find rows where revenue exceeds the upper bound
high_revenue_outliers = dataset[dataset['revenue'] > upper_bound]
print(f"Number of high-revenue outliers: {len(high_revenue_outliers)}")
high_revenue_outliers.sort_values(by='revenue', ascending=False).head()
```

visulize with a box plot

```python
sns.boxplot(data=dataset, x='revenue', color='lightcoral')
plt.title("Revenue Distribution with Outliers")
plt.show()
```
